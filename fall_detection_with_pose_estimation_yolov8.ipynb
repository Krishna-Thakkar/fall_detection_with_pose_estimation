```python

cap = cv2.VideoCapture('/home/mind/projects/projects/pose_estimation/videos/girl_falling.mp4')
NoneType = type(None)
if (cap.isOpened() == False):
    print('Error while trying to read video. Please check path again')

frame_width = 1920

frame_height = 1080

vid_write_image = letterbox(cap.read()[1], (frame_width), stride=64, auto=True)[0]
resize_height, resize_width = vid_write_image.shape[:2]
# out = cv2.VideoWriter('/home/mind/projects/projects/pose_estimation/output-N1IQOkcGESotWAcMVBIUOy5GQiFbDgtqYmMmEn8wMRcccisbdVlHaGEETGA(online-video-cutter.com).avi',
#                     cv2.VideoWriter_fourcc(*'XVID'), int(cap.get(5)),
#                     (resize_width, resize_height))

video_writer = cv2.VideoWriter("/home/mind/projects/projects/pose_estimation/output_girl_falling.mp4",                               
                               cv2.VideoWriter_fourcc(*'mp4v'),
                               int(cap.get(5)),
                               (frame_width, frame_height))

img = cv2.imread('/home/mind/projects/projects/pose_estimation/warning.jpg') 
new_image = cv2.resize(img, (250, 100))

frame_count = 0
flag = False
fps = int(cap.get(5))

while(cap.isOpened):

    ret, frame = cap.read()

    if ret:
        frame = cv2.resize(frame, (1920, 1080))
        # print(flag)
        frame_count += 1 
        
        tracks = model.track(frame.copy(), persist=True, show=False, classes=[0], verbose=False, conf=0.1)
        # print(len(tracks))
        if isinstance(tracks[0].boxes.id, NoneType):
            pass
        else:
            track_ids = tracks[0].boxes.id.int().cpu().tolist()
            # print(track_ids)
            results = model(source=frame.copy(),
                    show=False,
                    conf=0.3,
                    save=False,
                    verbose=False,
                    stream=True,
            )  
            # print(results[0]) 
            
            for result in results:
                # plt.imshow(result.orig_img) 
                # plt.show()
                for i in range(len(result.boxes.xywh)):
                    width = result.boxes.xywh[i][2]
                    height = result.boxes.xywh[i][3]
                    # print(result.boxes.xywh[i][0], result.boxes.xywh[i][1])
                    # print('width', width)
                    # print('height', height)
                    
                    if height < 1.5*width:
                        # print(i)
                        # print('Person has fallen down.')
                        flag = True
                    
            if flag:
                # p1 = (420, 75)
                # p2 = (633, 105)  
                # frame = cv2.rectangle(frame.copy(), p1, p2, (255, 255, 255), -1)
                # frame = cv2.putText(frame.copy(), f'Fall detected', (425, 100), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=1.5, color=[0, 0, 255], thickness=2, lineType=cv2.LINE_AA)
                if frame_count <= fps//2:
                    frame[100:200, 850:1100] = cv2.addWeighted(frame[100:200, 850:1100], 0, new_image, 1, 0)
                else:
                    frame[100:200, 850:1100] = cv2.addWeighted(frame[100:200, 850:1100], 1, new_image, 0, 0)
            
                            # cv2.putText(frame, f'Person {track_ids[i]} has fallen down.', (int(result.boxes.xywh[i][0]), int(result.boxes.xywh[i][1])), 0, 0.5, [0, 0, 255], thickness=2, lineType=cv2.LINE_AA)
                        # flag = False
            
            # new_results = model(source=frame,
            #             show=True,
            #             conf=0.3,
            #             save=True,
            #             verbose=True,
            #             stream=True,
            # )
            # print(out)
            # out.write(frame)   
            cv2.imshow('video', frame)
            if cv2.waitKey(1) & 0xFF == ord('q'): 
                break
        if frame_count == fps:
                frame_count = 0
        video_writer.write(frame) 
    else:
        break
    
video_writer.release()
cap.release()
cv2.destroyAllWindows() 
```
